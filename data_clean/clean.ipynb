### Data Cleaning

```python
# Convert date columns to datetime format
date_cols = ["declarationDate", "incidentBeginDate", "incidentEndDate"]

for col in date_cols:
    declarations[col] = pd.to_datetime(declarations[col], errors='coerce')
```

```python
# Ensure numeric columns are number types
num_cols_summ = [
    "totalAmountIhpApproved",
    "totalAmountHaApproved",
    "totalAmountOnaApproved",
    "totalObligatedAmountPa",
    "totalObligatedAmountHmgp"
]

for col in num_cols_summ:
    dis_summ[col] = pd.to_numeric(dis_summ[col], errors='coerce')

pub_assistance["totalObligated"] = pd.to_numeric(pub_assistance["totalObligated"], errors='coerce')
```

```python
(pub_assistance["totalObligated"] == 0).sum()
```

As earlier determined, no null values in pub_assistance and no project had totalObligated from government as zero. totalObligated > 0 always. This means there are no invalid projects.

### Feature Engineering

```python
# engineer time-based features from declarations dataset
decl = declarations.copy()

decl["declaration_year"] = decl["declarationDate"].dt.year
decl["declaration_month"] = decl["declarationDate"].dt.month
decl["declaration_quarter"] = decl["declarationDate"].dt.quarter

decl["incident_duration_days"] = (
    decl["incidentEndDate"] - decl["incidentBeginDate"]
).dt.days

decl["declaration_lag_days"] = (
    decl["declarationDate"] - decl["incidentBeginDate"]
).dt.days

decl["incident_open_flag"] = decl["incidentEndDate"].isna().astype(int)

decl.head()
```

```python
# engineer target variable in disaster summaries dataset
summ = dis_summ.copy()

summ["target_pa_cost"] = summ["totalObligatedAmountPa"].fillna(0)

summ["target_total_recovery_cost"] = (
    summ["totalAmountIhpApproved"].fillna(0)
    + summ["totalObligatedAmountPa"].fillna(0)
    + summ["totalObligatedAmountHmgp"].fillna(0)
)

# Log transform (to avoid skewness and handle zero values)
summ["log_target_pa_cost"] = np.log1p(summ["target_pa_cost"])
summ["log_target_total_cost"] = np.log1p(summ["target_total_recovery_cost"])

summ.head()
```

```python
# engineer features from public assistance dataset to prepare for merging with disaster summaries and declarations dataset
pa = pub_assistance.copy()

pa_agg = (
    pa.groupby("disasterNumber")
    .agg(
        pa_project_count=("pwNumber", "count"),
        pa_total_obligated_sum=("totalObligated", "sum"),
        pa_total_obligated_mean=("totalObligated", "mean"),
        pa_total_obligated_median=("totalObligated", "median"),
        pa_total_obligated_max=("totalObligated", "max")
    )
    .reset_index()
)

print(pa_agg.shape[0])
pa_agg.head()
```

```python
# check project size features
size_counts = (
    pa.pivot_table(
        index="disasterNumber",
        columns="projectSize",
        values="pwNumber",
        aggfunc="count",
        fill_value=0
    )
    .reset_index()
)

size_counts.columns = [str(c).lower() for c in size_counts.columns]

if "large" in size_counts.columns and "small" in size_counts.columns:
    size_counts["large_project_ratio"] = (
        size_counts["large"] / (size_counts["small"] + size_counts["large"])
    )
size_counts = size_counts.rename(columns={"disasternumber": "disasterNumber"})
size_counts.head()
```

```python
# check disasters with high ratio of large projects
print(size_counts[size_counts['large_project_ratio'] > 0.5].shape[0])
size_counts[size_counts['large_project_ratio'] > 0.5].head(10)
```

There have been 34 large scale disaster recovery projects since 2010.

```python
# engineer feature to show state-level disaster frequency
state_freq = (
    decl.groupby("state")["disasterNumber"]
    .count()
    .rename("state_disaster_frequency")
    .reset_index()
)

decl = decl.merge(state_freq, on="state", how="left")

decl.head()
```

### Merge

```python
dis_fact = (
    decl
    .merge(summ, on="disasterNumber", how="left")
    .merge(pa_agg, on="disasterNumber", how="left")
    .merge(size_counts, on="disasterNumber", how="left")
)

dis_fact.head()
```

```python
# create missing value indicators and fill missing values with zeros for funding feature columns - useful for model learning
summary_cols = [
    "totalAmountIhpApproved",
    "totalAmountHaApproved",
    "totalAmountOnaApproved",
    "totalObligatedAmountPa",
    "totalObligatedAmountHmgp"
]

pa_feature_cols = [
    "pa_project_count",
    "pa_total_obligated_sum",
    "pa_total_obligated_mean",
    "pa_total_obligated_median",
    "pa_total_obligated_max",
    "large",
    "small",
    "large_project_ratio"
]

for col in summary_cols:
    dis_fact[f"{col}_missing"] = dis_fact[col].isna().astype(int)

dis_fact[summary_cols] = dis_fact[summary_cols].fillna(0)
dis_fact[pa_feature_cols] = dis_fact[pa_feature_cols].fillna(0)

dis_fact.head()
```

```python
print(dis_fact.isnull().mean().sort_values(ascending=False).head(20))
```

```python
# remove rows with missing target variable values
dis_fact = dis_fact[
    (dis_fact["target_total_recovery_cost"].notna()) &
    (dis_fact["target_pa_cost"].notna())
].copy()

print(dis_fact.shape[0])
```
